# Описание задачи

Электрокардиография (ЭКГ) - ключевой диагностический инструмент для оценки состояния сердца пациента. Необходимо по предоставленным данным определить, есть ли у пациента заболевание сердца или нет, и, если есть, то какое.

## Структура датасета





Набор данных включает 21837 клинических записей ЭКГ в 12 отведениях продолжительностью 10 секунд от 18885 пациентов, из которых 52% - мужчины и 48% - женщины, возраст которых охватывает весь диапазон от 0 до 95 лет. 

Файлы сигналов хранятся в формате WaveForm DataBase (WFDB) с точностью до 16 бит и частотой дискретизации 500 Гц (records500 /). Для удобства пользователя также есть субдискретизированные версии данных сигнала с частотой дискретизации 100 Гц (records100 /). 

Все соответствующие метаданные хранятся в ptbxl_database.csv. Он содержит 28 столбцов, которые можно разделить на следующие категории:

1. Идентификаторы: каждая запись идентифицируется уникальным ecg_id. Соответствующий пациент кодируется через Patient_id. Пути к исходной записи (500 Гц) и субдискретизированной версии записи (100 Гц) хранятся в filename_hr и filename_lr.
2. Общие метаданные: возраст, пол, рост, вес, медсестра, сайт, устройство и дата записи.
3. Операторы ЭКГ: основными компонентами являются scp_codes и report (строка отчета). Дополнительные поля: heart_axis (ось сердца), infarction_stadium1, infarction_stadium2, validated_by, second_opinion, initial_autogenerated_report и validated_by_human.
4. Метаданные сигнала: качество сигнала, такое как шум (static_noise и burst_noise), отклонения базовой линии (baseline_drift) и другие артефакты, такие как electrodes_problems.


В файле scp_statements.csv хранятся данные, соответствующие отметкам scp_codes.
