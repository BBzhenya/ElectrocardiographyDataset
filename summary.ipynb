В данной работе была рассмотрена задача классфикации, в которой необходимо по данным ЭКГ предсказать группу (класс) сердечных заболеваний.

В работе были рассмотрены 4 метода машинного обучения - метод k ближайших соседей, метод опорных векторов, лиенйный дискриминантный анализ и нейронные сети.
Ни один из методов не дал хороших результатов на данном датасете.
Лучше всего себя показал метод опорных векторов с полиномиальным ядром. Точность вычислений на обучающей выборке 56%. 
Мы часто сталкивались с недообучением или переобучением. При использовании любого из методов ошибка на тестовой выборке была 70-80% (кроме SVM, как говорилось ранее). 

Данные в датасете не были сбалансированы, поэтому к обучающей выборке был применён метод, который борется с несбалансрованностью классов. Данный метод удаляет представителей классов, которые преоладают. После приаенения данного метода размер обучающей выборки уменьшился в 2,5 раза. 
Одна из причин плохих результатов работы, возможно, является маленькая обучающая выборка.

Ещё одна проблема, с которой мы столкнулись - большие временные затраты. Уже с такой маленькой обучающей выборкой некоторые методы работали более 2-х часов.

Возможной причиной плохих результатов может являться неправильная подборка гиперпараметров. Возможно, при каких-то иных их комбинациях результаты получились бы лучше.

Данные не были нормализированы или стандартизированы. Это тоже может влиять на результат.

P.S. Это не вошло в итоговую работу, но на некоторых алгоритмах были испробованы нормализованные/стандартизированные данные, это не привело к успеху.

Если почитать литературу на данную тему, то можно найти, что данную задачу лучше всего решать с помощью глубокого обучения. С помощью этого метода была достигнута высокая точность в классификации сердечных заболеваний по ЭКГ.
